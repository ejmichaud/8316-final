```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
sizes = [4, 6, 8, 12, 16, 24, 32, 48, 64, 128]
temperatures = np.linspace(1, 3.5, 50)
```

```python
for n, i in enumerate(range(0, 50, 9)):
    plt.subplot(1, 6, n+1)
    N = 32
    T = temperatures[i]
    nsim = 2000
    data = torch.load(f"/om/user/ericjm/results/class/8.316/final/all-0/{N}_{T}_{nsim}.pt")
    plt.imshow(data['data'][0], interpolation='none')
    plt.title(f"T={T:.2f}", fontsize=10)
    plt.axis('off')
plt.tight_layout(pad=0.3)
plt.savefig('../figures/grids32x32-6temps.png', dpi=400)
```

```python
plt.figure(figsize=(8, 8))
n = 1
for j in range(8): # sizes
    for i in range(0, 50, 7): # temperatures
        plt.subplot(8, 8, n)
        N = sizes[j]
        T = temperatures[i]
        nsim = 2000
        data = torch.load(f"/om/user/ericjm/results/class/8.316/final/all-0/{N}_{T}_{nsim}.pt")
        plt.imshow(data['data'][0], interpolation='none')
        # plt.title(f"T={T:.2f}", fontsize=10)
        plt.axis('off')
        n += 1
plt.tight_layout(pad=0.3)
plt.savefig('../figures/grids-gridsearch.png', dpi=300)
```
